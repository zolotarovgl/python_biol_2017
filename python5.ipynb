```python
import pandas as pd
import numpy as np
```

```python
url = 'https://www.czso.cz/documents/10180/32853391/300002164403.xlsx/1fed9fb2-4dbf-4b0f-bfbe-4c2d3202a18f?version=1.1'
data = pd.read_excel(url,header = None, index_col = 0, skiprows = 6)
data
```

```python
col1 = data.iloc[0].fillna(method = 'ffill')
col3 = data.iloc[3].fillna(method = 'ffill')
col2 = data.iloc[2].fillna(method = 'ffill')
columns = [col1, col2, col3]

columns = list(zip(*columns))
```

```python
data.columns = columns
data = data.drop(data.index[[0,1,2]])
data
```

```python
data.columns = pd.MultiIndex.from_tuples(data.columns)
```
